# Filtrando Data Frame

```python
# Selecionar colunas por índice
indices_colunas = [1, 2, 3, 8, 10, 11, 12, 13, 14, 17, 18, 20, 21, 22, 23, 25, 27]

df_filtrado = df.iloc[:, indices_colunas]

df_filtrado
```

```python
df_filtrado.set_index(df_filtrado.columns[0], inplace=True)
df_filtrado
```

```python
col_13 = df_filtrado.columns[13]  
col_14 = df_filtrado.columns[14]  

new_column_order = [col_13, col_14] + [col for i, col in enumerate(df_filtrado.columns) if i not in [13, 14]]

df_filtrado = df_filtrado[new_column_order]

df_filtrado
```

```python
# Analisando os valores nulos
df_filtrado.isnull().sum()
```

```python
df_filtrado.describe()
```

# Data Frame com valores numéricos

```python
df_filtrado_numerico = df_filtrado.select_dtypes(include=[int, float])
df_filtrado_numerico
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

correlacao = df_filtrado_numerico.corr() 
#visualização das correlações
cores = sns.color_palette('light:salmon', as_cmap=True)
mask = np.zeros_like(correlacao)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(22, 15))
    ax = sns.heatmap(correlacao, cmap=cores, mask=mask, square=True, fmt='.2f', annot=True)
```

```python
sns.set(style="whitegrid")

for col in df_filtrado_numerico.columns:
    plt.figure(figsize=(12, 8))  
    sns.boxplot(y=df_filtrado_numerico[col])
    plt.title(f'Boxplot da Coluna {col}', fontweight='bold', fontsize=16)
    plt.ylabel(col, fontsize=14)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Iterar por cada coluna no DataFrame
for col in df_filtrado_numerico.columns:
    plt.figure(figsize=(12, 8)) 
    plt.boxplot(df_filtrado_numerico[col].dropna())  # removendo os nan
    plt.title(f'Boxplot da Coluna {col}', fontweight='bold', fontsize=16)
    plt.ylabel('Valores', fontsize=14)
    plt.xlabel(col, fontsize=14)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.ylim([df_filtrado_numerico[col].min() - 1, df_filtrado_numerico[col].max() + 1])  
    plt.show()
```
